```python
1. What are the five key concepts of Object-Oriented Programming (OOP)?
1. Class and Object
class Animal:
    def __init__(self, name, species):
        self.name = name  # Object's attribute
        self.species = species

    def speak(self):
        relative:
        return f"{self.name} makes a sound."

# Create an object (instance) of class Animal
dog = Animal("Buddy", "Dog")
print(dog.speak())
2. Encapsulation
class Animal:
    def __init__(self, name, species):
        self._name = name  # Protected attribute (single underscore)
        self._species = species

    def speak(self):
        return f"{self._name} makes a sound."

    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

# Using getter and setter to access and modify attributes
cat = Animal("Whiskers", "Cat")
print(cat.get_name())
cat.set_name("Mittens")
print(cat.get_name())
3. Inheritance
class Dog(Animal):  # Dog class inherits from Animal
    def speak(self):
        return f"{self._name} barks."

dog = Dog("Buddy", "Dog")
print(dog.speak())
4. Polymorphism
class Cat(Animal):
    def speak(self):
        return f"{self._name} meows."

animals = [Dog("Buddy", "Dog"), Cat("Whiskers", "Cat")]

for animal in animals:
    print(animal.speak())  # Same method name 'speak', different behaviors
5. Abstraction
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Instantiating Dog and Cat objects
dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())

2. Write a Python class for a `Car` with attributes for `make`, `model`, and `year`. Include a method to display the car's information.
class Car:
    def __init__(self, make, model, year):
        self.make = make    # Car's make (e.g., Toyota)
        self.model = Ordinary Car or Electric
        self.year = make202#
        
  def Displayclass __init__(

3. Explain the difference between instance methods and class methods. Provide an example of each.
Key Differences:
Instance methods:
Belong to the instance of a class.
Can access and modify instance-level data (attributes).
Take self as the first parameter, which represents the instance calling the method.
Class methods:
Belong to the class itself rather than the instance.
Can access class-level data (attributes shared across all instances).
Take cls as the first parameter, which represents the class calling the method.
Declared with the @classmethod decorator.
Example:
 class Car:
    # Class attribute (shared across all instances)
    number_of_wheels = 4

    def __init__(self, make, model, year):
        # Instance attributes (unique to each instance)
        self.make = make
        self.model = model
        self.year = year

    # Instance method (works with instance attributes)
    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

    # Class method (works with class attributes)
    @classmethod
    def get_wheel_count(cls):
        return f"All cars have {cls.number_of_wheels} wheels."

# Creating an instance of Car
my_car = Car("Toyota", "Corolla", 2020)

# Calling the instance method
print(my_car.display_info())  # Output: 2020 Toyota Corolla

# Calling the class method (through instance or class)
print(Car.get_wheel_count())  # Output: All cars have 4 wheels
print(my_car.get_wheel_count())  # Output: All cars have 4 wheels

 4. How does Python implement method overloading? Give an example.
 Python does not support traditional method overloading (as seen in languages like Java or C++) where multiple methods with the same name but different argument types or counts are defined. Instead, Python achieves this using default arguments, variable-length arguments (*args and **kwargs), or by explicitly checking the types and number of arguments inside the method.
 class MathOperations:
    def add(self, *args):
        # If one argument is passed
        if len(args) == 1:
            return args[0] + 10  # Adds 10 if only one number is passed
        # If two arguments are passed
        elif len(args) == 2:
            return args[0] + args[1]
        # If three arguments are passed
        elif len(args) == 3:
            return args[0] + args[1] + args[2]
        else:
            return "Invalid number of arguments"

# Creating an object
math_op = MathOperations()

# Calling 'add' with different numbers of arguments
print(math_op.add(5))        
print(math_op.add(5, 3))       
print(math_op.add(5, 3, 2))    

5. What are the three types of access modifiers in Python? How are they denoted?
 1. Public (No leading underscore)
class Car:
    def __init__(self, make, model):
        self.make = make    # Public attribute
        self.model = model  # Public attribute

    def display_info(self):
        return f"Car: {self.make} {self.model}"  # Public method

# Accessing public attributes and methods
car = Car("Toyota", "Camry")
print(car.make)  # Output: Toyota
print(car.display_info())  # Output: Car: Toyota Camry
2. Protected (Single leading underscore _)
class Car:
    def __init__(self, make, model):
        self._make = make   # Protected attribute
        self._model = model # Protected attribute

    def _display_info(self):  # Protected method
        return f"Car: {self._make} {self._model}"

# Accessing protected attributes and methods (not recommended, but possible)
car = Car("Honda", "Civic")
print(car._make)  # Output: Honda (accessible but should not be accessed)
print(car._display_info())  # Output: Car: Honda Civic (not recommended)
3. Private (Double leading underscore __)
class Car:
    def __init__(self, make, model):
        self.__make = make  # Private attribute
        self.__model = model  # Private attribute

    def __display_info(self):  # Private method
        return f"Car: {self.__make} {self.__model}"

    def get_car_info(self):  # Public method to access private data
        return self.__display_info()

# Accessing private attributes and methods
car = Car("Ford", "Fusion")
# print(car.__make)  # This will raise an AttributeError
# print(car.__display_info())  # This will also raise an AttributeError

# Accessing private attributes through name mangling (not recommended)
print(car._Car__make)  # Output: Ford (using name mangling)
print(car.get_car_info())  # Output: Car: Ford Fusion

 6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.
 1. Single Inheritance
class Parent:
    def greet(self):
        return "Hello from Parent"

class Child(Parent):
    pass

child = Child()
print(child.greet())  # Output: Hello from Parent
2. Multiple Inheritance
class Parent1:
    def greet(self):
        return "Hello from Parent1"

class Parent2:
    def greet_again(self):
        return "Hello from Parent2"

class Child(Parent1, Parent2):
    pass

child = Child()
print(child.greet())        # Output: Hello from Parent1
print(child.greet_again())  # Output: Hello from Parent2
3. Multilevel Inheritance
class Grandparent:
    def greet(self):
        return "Hello from Grandparent"

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

child = Child()
print(child.greet())  # Output: Hello from Grandparent
4. Hierarchical Inheritance
class Parent:
    def greet(self):
        return "Hello from Parent"

class Child1(Parent):
    pass

class Child2(Parent):
    pass

child1 = Child1()
child2 = Child2()
print(child1.greet())  # Output: Hello from Parent
print(child2.greet())  # Output: Hello from Parent
5. Hybrid Inheritance
class Parent:
    def greet(self):
        return "Hello from Parent"

class Child1(Parent):
    pass

class Child2(Parent):
    pass

class Grandchild(Child1, Child2):
    pass

grandchild = Grandchild()
print(grandchild.greet())  # Output: Hello from Parent

 7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?
 The Method Resolution Order (MRO) is the order in which Python looks for a method or attribute in a hierarchy of classes during inheritance. This is particularly important in the case of multiple inheritance to determine the order in which parent classes are searched when calling a method.
Example of Multiple Inheritance and MRO:
class A:
    def show(self):
        return "Method in A"

class B(A):
    def show(self):
        return "Method in B"

class C(A):
    def show(self):
        return "Method in C"

class D(B, C):
    pass  # D inherits from both B and C

# Instantiate class D
d = D()

# Access the method from the MRO
print(d.show())  # Output: Method in B

# Retrieve the MRO using __mro__
print(D.__mro__)

# Retrieve the MRO using the mro() method
print(D.mro())

 8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses `Circle` and `Rectangle` that implement the `area()` method.
from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    # Abstract method
    @abstractmethod
    def area(self):
        pass

# Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    # Implementing the abstract method
    def area(self):
        return math.pi * self.radius ** 2

# Subclass: Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    # Implementing the abstract method
    def area(self):
        return self.width * self.height

# Instantiate objects and calculate area
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of Circle: {circle.area():.2f}")  # Output: Area of Circle: 78.54
print(f"Area of Rectangle: {rectangle.area()}")  # Output: Area of Rectangle: 24

 9. Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.
 from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2

# Subclass: Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

# Polymorphic function to calculate and print area
def print_area(shape):
    print(f"The area of the {shape.__class__.__name__} is {shape.area():.2f}")

# Instantiate objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Call the polymorphic function with different shape objects
print_area(circle)      # Output: The area of the Circle is 78.54
print_area(rectangle)   # Output: The area of the Rectangle is 24.00

 10. Implement encapsulation in a `BankAccount` class with private attributes for `balance` and `account_number`. Include methods for deposit, withdrawal, and balance inquiry.
 class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self.__balance = initial_balance        # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            print("Insufficient funds.")

    def get_balance(self):
        return f"Current balance: ${self.__balance:.2f}"

    def get_account_number(self):
        return f"Account number: {self.__account_number}"

# Example usage
account = BankAccount("123456789", 1000)

print(account.get_account_number())  # Output: Account number: 123456789
print(account.get_balance())         # Output: Current balance: $1000.00

account.deposit(500)                  # Output: Deposited: $500.00
print(account.get_balance())         # Output: Current balance: $1500.00
 

account.withdraw(200)                 # Output: Withdrew: $200.00
print(account.get_balance())         # Output: Current balance: $1300.00

account.withdraw(1500)                # Output: Insufficient funds.
print(account.get_balance())         # Output: Current balance: $1300.00
 
 11. Write a class that overrides the `__str__` and `__add__` magic methods. What will these methods allow you to do?
 class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        # This method defines the string representation of the object
        return f"Point({self.x}, {self.y})"

    def __add__(self, other):
        # This method defines how to add two Point objects
        if isinstance(other, Point):
            return Point(self.x + other.x, self.y + other.y)
        return NotImplemented

# Example usage
point1 = Point(1, 2)
point2 = Point(3, 4)

# Using the __str__ method
print(point1)  # Output: Point(1, 2)
print(point2)  # Output: Point(3, 4)

# Using the __add__ method
point3 = point1 + point2
print(point3)  # Output: Point(4, 6)

 12. Create a decorator that measures and prints the execution time of a function.
 import time

def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record start time
        result = func(*args, **kwargs)  # Call the actual function
        end_time = time.time()  # Record end time
        elapsed_time = end_time - start_time  # Calculate elapsed time
        print(f"Execution time of {func.__name__}: {elapsed_time:.4f} seconds")
        return result
    return wrapper

# Example usage
@timing_decorator
def slow_function(seconds):
    """Function that sleeps for a given number of seconds"""
    time.sleep(seconds)
    return "Finished"

# Call the decorated function
print(slow_function(2))  # Output will include the execution time

 13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
 class A:
    def show(self):
        print("Method in A")

class B(A):
    def show(self):
        print("Method in B")

class C(A):
    def show(self):
        print("Method in C")

class D(B, C):
    pass

# Create an instance of D
d = D()
d.show()  # Output: Method in B

 
14. Write a class method that keeps track of the number of instances created from a class.
 class InstanceCounter:
    # Class variable to keep track of the number of instances
    _instance_count = 0

    def __init__(self):
        # Increment the instance count every time a new instance is created
        InstanceCounter._instance_count += 1

    @classmethod
    def get_instance_count(cls):
        # Return the number of instances created
        return cls._instance_count

# Example usage
a = InstanceCounter()
b = InstanceCounter()
c = InstanceCounter()

# Get the number of instances created
print(f"Number of instances created: {InstanceCounter.get_instance_count()}")  # Output: Number of instances created: 3

 15. Implement a static method in a class that checks if a given year is a leap year.
 class YearUtils:
    @staticmethod
    def is_leap_year(year):
        """Check if the given year is a leap year."""
        # Leap year is divisible by 4
        # but not divisible by 100 unless also divisible by 400
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        else:
            return False

# Example usage
print(YearUtils.is_leap_year(2024))  # Output: True (2024 is a leap year)
print(YearUtils.is_leap_year(2023))  # Output: False (2023 is not a leap year)
print(YearUtils.is_leap_year(2000))  # Output: True (2000 is a leap year)
print(YearUtils.is_leap_year(1900))  # Output: False (1900 is not a leap year)




```
